```python
# Merging
order_items.merge(right=products,
                  how="left",
                  on="product_id")
```

```python
#Select columns and assign new variable
order_products = order_items.merge(right=products,
                  how="left",
                  on="product_id")[["order_id",
                                    "order_item_id",
                                    "product_id",
                                    "price",
                                    "product_category_name"]]

order_products
```

```python
#group by, with now using count and sum
# Aggregation required for grouping, or else redundant (in general)
per_category = order_products.groupby("product_category_name").agg({"price":"sum","order_id":"count"})

```

```python
# Combine it into one query, via method chaining

(order_items
.merge(products, how="left",
                 on="product_id")
.groupby("product_category_name")
.agg({"price":"sum",
      "order_id":"count"})
.reset_index())
```

```python
# Add column via assign

(order_items
.merge(products, how="left",
                 on="product_id")
.groupby("product_category_name")
.agg({"price":"sum",
      "order_id":"count"})
.reset_index()
.assign(newcol=1))
```

```python
# rename a column from select via .rename(columns={"colold":"colnew"})

(order_items
.merge(products, how="left",
                 on="product_id")
.groupby("product_category_name")
.agg({"price":"sum",
      "order_id":"count"})
.reset_index()
.rename(columns={"price":"revenue",
                 "order_id":"items_sold"})
.assign(newcol=1))
```

```python
# Sorting - or ORder by

(order_items
.merge(products, how="left",
                 on="product_id")
.groupby("product_category_name")
.agg({"price":"sum",
      "order_id":"count"})
.reset_index()
.rename(columns={"price":"revenue",
                 "order_id":"items_sold"})
.sort_values("revenue", ascending=False)
.head())
```
