#to combine multiple mappings via conditions:

```python
df["orange_density"].lt(.5) & df["kiwi_density"].gt(.5)
```

```python
# This returns a boolean map, where BOTH conditions are true

df[df["orange_density"].lt(.5) & df["kiwi_density"].gt(.5)]
```

```python
# Confusing: When does Panda know when you want to subset a column or a row? Dont reall know what this means..hmm

# to solve:
#use .loc[] or iloc[]

# df.loc[rows, columns]

df.loc[df["orange_density"].lt(.5), ["orange_mass"]]
```

```python
# or if you only want to subset columns (remove values lower than 0.5 for orange density):

df.loc[df["orange_density"].lt(.5),]
```

```python
# or only show the column
df.loc[:,df["kiwi_density"].gt(.5)]
```
